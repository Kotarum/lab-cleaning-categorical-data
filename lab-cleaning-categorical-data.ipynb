```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv('/home/kotarum/Ironhack/unit_4/lab-cleaning-categorical-data/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')  # Adjust the path as necessary
```

```python
customer_df
```

```python
customer_df.columns = [column.lower().replace(" ", "_") for column in customer_df.columns]
```

```python
customer_df
```

```python
numerical_df = customer_df.select_dtypes(include=['int64', 'float64'])  # Adjust types as necessary
categorical_df = customer_df.select_dtypes(include=['object'])
```

```python
for column in categorical_df.columns:
    plt.figure(figsize=(10, 5))
    sns.barplot(x=column, y='total_claim_amount', data=customer_df)  # Adjust the target variable as necessary
    plt.title(f'Mean Total Claim Amount by {column}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
for column in categorical_df.columns:
    print(f'Value counts for {column}:')
    print(categorical_df[column].value_counts())
    print('\n')
```
